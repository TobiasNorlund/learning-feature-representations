```python
from main import train_nce, get_mnist_data
import numpy as np
import scipy
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
minst_train, mnist_test, mnist_mean = get_mnist_data("../../data", add_noise=True)
```

```python
precision_matrix = train_nce(batch_size=100, eta=0.5, mnist_data_path="../../data/", max_steps=50)
precision_matrix = precision_matrix.detach().numpy()
```

```python
L = np.linalg.cholesky(precision_matrix)
```

```python
sample_image = mnist_mean + scipy.linalg.solve_triangular(L, np.random.normal(size=28*28), lower=True)
sample_image = sample_image.reshape(28, 28)
plt.imshow(sample_image)
```
